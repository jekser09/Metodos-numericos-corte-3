```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
#Cedula: 1018506713

a=-1
b=2
n=6

dx=(b-a)/n

xi = np.zeros(n+1)

for i in range(n+1):
  xi[i] = a+i*dx
print(f"Los extremos de los subintervalos son: \n {xi}\n ")

suma=0
for i in range(n-1):
  suma = suma + 1/xi[i+1]
aprox = (dx/2)*(2*suma + (-1/xi[0]) + (-1/xi[n]))
print(f"Valor aproximado del área es:\n {round(aprox,6)}\n")

x = sp.Symbol("x")
funcion=sp.exp(x+sp.cos(x))
valor = sp.integrate(funcion,(x,-1,2))
valor = float(valor)
print(f"Valor exacto:\n {round(valor,6)}\n")

error = (np.abs(valor - aprox)/valor)*100
print(f"Error relativo:\n {round(error,3)}%\n")

figura, ejes = plt.subplots()
x = np.linspace(-3,3,100)
x1=np.linspace(-1,6,100)
y = np.exp(x+np.cos(x))
ejes.plot(x,y, lw=1.5, color="#2EFF00")
ejes.grid()
ejes.set_title(r"-----$f(x)=e^{x+\cos{x}}$-----")
ejes.set_xlabel("X")
ejes.set_ylabel("Y")
ejes.axhline(0, color="#000000")
ejes.fill_between(x,0,y,color="#12C553",alpha=0.4)
```
