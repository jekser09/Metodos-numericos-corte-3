Una persona compra tres suplementos vitamínicos para enriquecer su dieta.
Para ello compra Ensure, Milo y Kola Granulada. La persona necesita 940 unidades diarias de fósforo, 670 unidades de zinc y 330 unidades de calcio. Se sabe que el Ensure tiene 50 unidades de fósforo por cada gramo, 30 unidades de zinc por cada gramo y 10 unidades de calcio por cada gramo. El milo 40 unidades de fósforo por cada gramo, 35 unidades de zinc por cada gramo y 25 unidades de calcio por cada gramo. Y la Kola granulada 20 unidades de fósforo por cada gramo, 15 unidades de zinc por cada gramo y 5 unidades de calcio por cada
gramo. Encontrar la cantidad de gramos por cada producto que debe consumir la persona para satisfacersus necesidades diarias. Plantear un sistema de ecuaciones y resolverlo empleando Python.

$x = Ensure$

$y = Milo$

$z = Kola Granulada$

\begin{matrix}
50x+40y+20z=940\\
30x+35y+15z=670\\
10x+25y+5z=330
\end{matrix}

La velocidad del agua que pasa desde un tanque cilíndrico hacia una tubería de longitud L en su parte
inferior está dada por la expresión:
$v=\sqrt{2gH}\tanh({\frac{\sqrt{2gH}}{2L}t)}$

Donde  $g = 9.8m/s2$, H es la altura inicial del agua (m), L es la longitud del tubo (m) y t el tiempo transcurrido (s). Halle la altura (H) necesaria para lograr una velocidad de $v = 4m/s$ en $3s$ para una tubería de $5m$. Por el método de bisección en el intervalo [0, 2] con 10 iteraciones y por el método de Newton con 10 iteraciones.

#Biseccion

```python
g=9.8
L=5
t=3

a=0
b=2

f_inf=4-np.sqrt(2*g*a)*np.tanh((np.sqrt(2*g*a)/(2*L))*t)
f_sup=4-np.sqrt(2*g*b)*np.tanh((np.sqrt(2*g*b)/(2*L))*t)
if f_inf*f_sup>0:
  print("La funcion no tiene raices en ese intervalo")
elif f_inf*f_sup<0:  
  for i in range(10):    
    x0=(a+b)/2
    f0=4-np.sqrt(2*g*x0)*np.tanh((np.sqrt(2*g*x0)/(2*L))*t)
    if f_inf*f0<0:
      b=x0
    elif f_sup*f0<0:
      a=x0
    print("(",i+1,")\tx0={:.10f}".format(x0))
else:
  print("Funcion inferior * Funcion superior >= 0")
```

#Newton

```python

H0 = float(2)
iteraciones = 10
H = sp.Symbol("H")
funcion =4-sp.sqrt(2*g*H)*sp.tanh((sp.sqrt(2*g*H)/(2*L))*t)
derivada = funcion.diff(H, 1)
for i in range(iteraciones):
  Hn = H0 - (funcion.subs(H, H0))/(derivada.subs(H, H0))
  H0 = Hn
  print("i = ",i+1,", x0 = {:.6f}".format(H0))
```
