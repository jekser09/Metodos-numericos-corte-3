5. Aplique el método de la bisección para obtener la raíz con 10 cifras decimales de exactitud para la función $f(x) = \sqrt{x} − cos(x)$ en el intervalo [0, 1].

```python
a=0
b=1
f_inf=np.sqrt(a)-np.cos(a)
f_sup=np.sqrt(b)-np.cos(b)
if f_inf*f_sup>0:
  print("La funcion no tiene raices en ese intervalo")
elif f_inf*f_sup<0:  
  for i in range(10):    
    x0=(a+b)/2
    f0=np.sqrt(x0)-np.cos(x0)
    if f_inf*f0<0:
      b=x0
    elif f_sup*f0<0:
      a=x0
    print("(",i+1,")\tx0={:.10f}".format(x0))
else:
  print("que tas buscando bro")

figura, ejes=plt.subplots()
x=np.linspace(0,10,200)
y=np.sqrt(x)-np.cos(x)
ejes.plot(x,y,color="#060980",lw=2)
ejes.set_title(r"$f(x)=\sqrt{x}-\cos(x)$")
ejes.set_xlabel("X")
ejes.set_ylabel("Y")
ejes.axhline(0,color="black")
ejes.grid()
```

7. Aplique el método de Newton para obtener la solución de la ecuación: $ln (x − 1) + cos(x − 1) = 0$, en el
intervalo $1.3 ≤ x ≤ 2$ para 6 iteraciones.

```python
x = np.linspace(1.3,2,100)
y1 = np.log(x-1)+np.cos(x-1)
figura, ejes = plt.subplots()
ejes.plot(x,y1,color = "green", label = r"$f(x) = ln(x-1)+cos(x-1)$", lw =2.0)
ejes.set_title("Metodo de Newton")
ejes.set_xlabel("x")
ejes.set_ylabel("y")
ejes.legend()
ejes.grid()

x0 = float(2)
iteraciones = 6
x = sp.Symbol("x")
funcion = sp.log(x-1)+sp.cos(x-1)
derivada = funcion.diff(x, 1)
for i in range(iteraciones):
  xn = x0 - (funcion.subs(x, x0))/(derivada.subs(x, x0))
  x0 = xn
  print(f"{i+1} x0={x0}")
```

8. La suma de dos números es 20.Si a cada uno se agrega su raíz cuadrada el producto de las dos sumas es 155.55.

  Encuentre los dos números con el método de Newton y 4 iteraciones.

$x=20-y$

$(20-y+\sqrt{20-y})(y+\sqrt{y})=155$

$(20-y+\sqrt{20-y})(y+\sqrt{y})-155=0$


```python
x = np.linspace(0,20,200)
y1 = (20-x+np.sqrt(20-x))*(x+np.sqrt(x))-155
figura, ejes = plt.subplots()
ejes.plot(x,y1,color = "green", label = r"$f(x) = (20-y+\sqrt{20-y})(y+\sqrt{y})-155$", lw =2.0)
ejes.set_title("Metodo de Newton")
ejes.set_xlabel("x")
ejes.set_ylabel("y")
ejes.legend()
ejes.grid()

y0 = float(2)
iteraciones = 4
y = sp.Symbol("y")
funcion = (20-y+sp.sqrt(20-y))*(y+sp.sqrt(y))-155
derivada = funcion.diff(y, 1)
for i in range(iteraciones):
  yn = y0 - (funcion.subs(y, y0))/(derivada.subs(y, y0))
  y0 = yn
  print(f"{i+1} y0={y0}")
```
